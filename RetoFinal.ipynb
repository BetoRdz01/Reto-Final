```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
from IPython.display import display,HTML
```

```python
#Leo mi archivo.csv
data = pd.read_csv("CountryTable.csv")
data
```

```python
data["independence_year"]=data["independence_year"].fillna(1900)
data["life_expectancy"]=data["life_expectancy"].fillna(round(data["life_expectancy"].mean(),1))
data["gnp_old"]=data["gnp_old"].fillna(round(data["gnp_old"].median(),1))
data["local_name"]=data["local_name"].fillna("Inexistente")
data["head_of_state"]=data["head_of_state"].fillna("Desconocido")
data["capital"]=data["capital"].fillna(0)
data["code2"]=data["code2"].fillna("NA")
data
```

```python
#Verificamos que ya no tenemos datos nulos.
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Identificamos outliers
y=data["gnp_old"]
Limite_Superior= y.mean() + 2*y.std()
Limite_Inferior= y.mean() - 2*y.std()
outliers= data[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data["gnp_old"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.3*iqr
Limite_Inferior_iqr= percentile25 - 1.3*iqr
outliers_iqr= data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data["gnp_old"]) 
plt.title("Outliers de gnp_old")
plt.show() #dibujamos el diagrama
```

```python
#Buscamos la media de gnp_old
gnp_mean = data["gnp_old"].mean()
gnp_mean
```

```python
#Observamos los outliers de los datos con un límite de 3 desviaciones estándar.  
y=data["gnp_old"]
Limite_Superior4= y.mean() + 3*y.std()
Limite_Inferior4= y.mean() - 3*y.std()
outliers4= data[(y>Limite_Superior4)|(y<Limite_Inferior4)]
outliers4
```

```python
#Cambiamos los outliers por la media obtenida previamente.
data["gnp_old"] = np.where(((data["gnp_old"]>Limite_Superior4) | (data["gnp_old"]<Limite_Inferior4)),66731.07531380754,data["gnp_old"])
data["gnp_old"]
```

```python
#Realizamos diagrama de caja para observar los cambios en la variable
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data["gnp_old"]) 
plt.title("Outliers de gnp_old")
plt.show() #dibujamos el diagrama
#Observamos los cambios en los outliers. Los cuartiles se vuelven más amplios a simple vista. 
```
